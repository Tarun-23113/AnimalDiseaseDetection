```python
from torch.utils.data import Dataset
import torchvision.transforms as transforms
from PIL import Image
import os

class Animal_Dataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.image_paths = []
        self.labels = []
        self.label_map = {}

        for idx, disease_name in enumerate(os.listdir(root_dir)):
            disease_path = os.path.join(root_dir, disease_name)
            if os.path.isdir(disease_path):
                self.label_map[disease_name] = idx
                for img_name in os.listdir(disease_path):
                    img_path = os.path.join(disease_path, img_name)
                    if os.path.isfile(img_path) and img_path.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
                        self.image_paths.append(img_path)
                        self.labels.append(idx)

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        label = self.labels[idx]
        image = Image.open(img_path).convert('RGB')

        if self.transform:
            image = self.transform(image)

        return image, label

    def get_class_names(self):
        return list(self.label_map.keys())

# Dataset Making
train_data = Animal_Dataset(root_dir='/content/data',
                             transform=transforms.Compose([
                                 transforms.Resize((224, 224)),
                                 transforms.RandomHorizontalFlip(p=0.5),
                                 transforms.RandomRotation(degrees=15),
                                 transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
                                 transforms.ToTensor(),
                                 transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
                             ]))

test_data = Animal_Dataset(root_dir='/content/test_data',
                           transform=transforms.Compose([
                               transforms.Resize((224, 224)),
                               transforms.ToTensor(),
                               transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
                           ]))
```

```python
from torch.utils.data import DataLoader
train_dataloader = DataLoader(train_data, batch_size=15, shuffle=True)
test_dataloader = DataLoader(test_data, batch_size=6, shuffle=True)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class ADD_Model(nn.Module):
    def __init__(self, num_classes=7):
        super(ADD_Model, self).__init__()

        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(32)
        self.conv2 = nn.Conv2d(32, 32, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(32)

        self.conv3 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.bn3 = nn.BatchNorm2d(64)
        self.conv4 = nn.Conv2d(64, 64, kernel_size=3, padding=1)
        self.bn4 = nn.BatchNorm2d(64)

        self.conv5 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.bn5 = nn.BatchNorm2d(128)
        self.conv6 = nn.Conv2d(128, 128, kernel_size=3, padding=1)
        self.bn6 = nn.BatchNorm2d(128)

        self.pool = nn.MaxPool2d(2, 2)
        self.dropout = nn.Dropout(0.5)

        self.fc1_input_size = self._get_fc1_input_size()
        self.fc1 = nn.Linear(self.fc1_input_size, 512)
        self.fc2 = nn.Linear(512, num_classes)

    def _get_fc1_input_size(self):
        x = torch.zeros(1, 3, 224, 224)
        x = self.pool(F.leaky_relu(self.bn1(self.conv1(x))))
        x = self.pool(F.leaky_relu(self.bn2(self.conv2(x))))
        x = self.pool(F.leaky_relu(self.bn3(self.conv3(x))))
        x = self.pool(F.leaky_relu(self.bn4(self.conv4(x))))
        x = self.pool(F.leaky_relu(self.bn5(self.conv5(x))))
        x = self.pool(F.leaky_relu(self.bn6(self.conv6(x))))
        return x.numel()

    def forward(self, x):
        x = self.pool(F.leaky_relu(self.bn1(self.conv1(x))))
        x = self.pool(F.leaky_relu(self.bn2(self.conv2(x))))
        x = self.pool(F.leaky_relu(self.bn3(self.conv3(x))))
        x = self.pool(F.leaky_relu(self.bn4(self.conv4(x))))
        x = self.pool(F.leaky_relu(self.bn5(self.conv5(x))))
        x = self.pool(F.leaky_relu(self.bn6(self.conv6(x))))

        x = x.view(x.size(0), -1)
        x = F.leaky_relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x
```

```python
import torch.optim as optim

loss_fn = nn.CrossEntropyLoss()

optimizer = optim.Adam(model.parameters(), lr=0.0001)

scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=3, gamma=0.1)
```

```python
epochs = 5

for epoch in range(epochs):
    print(f"Epoch {epoch+1}/{epochs}")
    model.train()

    running_loss = 0.0
    correct_train = 0
    total_train = 0

    for X, y in train_dataloader:
        X, y = X.to(device), y.to(device)

        optimizer.zero_grad()
        outputs = model(X)

        loss = loss_fn(outputs, y)
        running_loss += loss.item()

        _, predicted_train = torch.max(outputs, 1)
        correct_train += (predicted_train == y).sum().item()
        total_train += y.size(0)

        loss.backward()
        optimizer.step()

    avg_train_loss = running_loss / len(train_dataloader)
    train_accuracy = 100 * correct_train / total_train
    print(f"Training Loss: {avg_train_loss:.4f}, Training Accuracy: {train_accuracy:.2f}%")

    model.eval()
    correct_test = 0
    total_test = 0
    running_test_loss = 0.0

    with torch.no_grad():
        for X, y in test_dataloader:
            X, y = X.to(device), y.to(device)
            outputs = model(X)
            loss = loss_fn(outputs, y)
            running_test_loss += loss.item()

            _, predicted_test = torch.max(outputs, 1)
            correct_test += (predicted_test == y).sum().item()
            total_test += y.size(0)

    avg_test_loss = running_test_loss / len(test_dataloader)
    test_accuracy = 100 * correct_test / total_test
    print(f"Test Loss: {avg_test_loss:.4f}, Test Accuracy: {test_accuracy:.2f}%")
```
